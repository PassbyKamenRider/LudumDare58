```python
import pandas as pd
```

```python
df_cards = pd.read_csv('card_properties.csv')
df_cards = df_cards[['Card Name', 'Class', 'Cost', 'Description', 'ATK', 'DEF', 'Rarity', 'CodeNum', 'Price']]
df_cards['Price'] = df_cards['Price'].fillna(0).astype(int)
df_cards['ATK'] = df_cards['ATK'].fillna(0).astype(int)
df_cards['DEF'] = df_cards['DEF'].fillna(0).astype(int)
df_cards
```

```python
df_cards.to_json('card_properties.json', orient='records', indent=4)
```
